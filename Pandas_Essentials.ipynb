<font color="red" size="4"><b>Grouping and Aggregate functions</b></font>


```python
# Applying groupby
data_by_location=weather.groupby(by=['Location']).mean()
data_by_location.head()
```

```python
data_by_location.sort_values(['Rainfall'],ascending=True).head()
```

The groupby() function returns a Pandas object, which can be used further to perform the desired aggregate functions.

```python
data_bymonth=weather.groupby(['Month']).mean()
data_bymonth.head()
```

```python
data_bymonth.sort_values(['MinTemp'],ascending=True)
```

```python
# Creating a 'chill factor' column
from math import sqrt
def wci(x):
  velocity=x['WindGustSpeed']
  mintemp=x['MinTemp']
  return (10*sqrt(velocity)-velocity+10.5)*(33-mintemp)
```

```python
weather['WCI']=weather.apply(wci,axis=1)
```

```python
weather.head()
```

<font color="red" size="4"><b>Data Merging Operations</b></font>


```python
tea_sales=pd.read_csv('/content/drive/MyDrive/New folder/junesales.csv')
tea_sales.head()
```

```python
tea_sales['Month']=pd.DatetimeIndex(tea_sales['Date']).month
tea_sales.head()
```

```python
weather.head()
```

```python
# Find total tea sales in the month of June for Newcastle region

newcastle_data=weather[(weather['Location']=='Newcastle') & (weather['Year']==2011) & (weather['Month']==6)]
newcastle_data.head()
```

```python
# df.merge merges two dataframes horizontally with a common column
merge_data=newcastle_data.merge(tea_sales,on='Month')
merge_data.head()
```

The attribute how n the code above specifies the type of merge that is to be performed. Merges are of the following different types:

    left: This will select the entries only in the first dataframe.
    right: This will consider the entries only in the second dataframe.
    outer: This takes the union of all the entries in the dataframes.
    inner: This will result in the intersection of the keys from both frames.

Concatenating dataframes
Concatenation is much more straightforward than merging. It is used when you have dataframes with the same columns and want to stack them on top of each other, or with the same rows and want to append them side by side.

You can add columns or rows from one dataframe to another using the concat function:

pd.concat([dataframe_1, dataframe_2], axis = _)


To append rows, you have to set the axis value as 0. For adding columns from one dataframe to another, the axis value must be set as 1. If there are any extra columns or rows where there are no values, they are replaced with NaN.
